```python
import pandas as pd
import numpy as np

train_df = pd.read_csv('data/train.csv')
```

```python
building_ids = train_df['건물번호'].unique()
```

```python
test_df = pd.read_csv('data/test.csv')
```

```python
# '일조(hr)', '일사(MJ/m2)', '전력소비량(kWh)' 열을 추가하고, 초기값을 NaN으로 설정합니다.
test_df['일조(hr)'] = np.nan
test_df['일사(MJ/m2)'] = np.nan
test_df['전력소비량(kWh)'] = np.nan
```

```python
test_df.head()
```

```python
new_test_df = pd.DataFrame()
```

```python
for building_id in building_ids:
    last_24_train = train_df[train_df['건물번호'] == building_id].tail(24)

    test_data_for_building = pd.concat([last_24_train, test_df[test_df['건물번호'] == building_id]], ignore_index=True)

    # 결과를 new_test_df에 추가합니다.
    new_test_df = pd.concat([new_test_df, test_data_for_building])
```

```python
new_test_df.head()
```

```python
new_test_df = new_test_df.sort_values(by=['건물번호', '일시'])
```

```python
new_test_df.head()
```

```python
new_test_df.to_csv('data/new_test.csv', index=False)
```

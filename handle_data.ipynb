```python
import numpy as np

def positional_encoding(seq_len, d_model):
    """
    seq_len: 시퀀스의 길이
    d_model: 특징 차원의 수
    """
    pe = np.zeros((seq_len, d_model))
    
    for pos in range(seq_len):
        for i in range(0, d_model, 2):
            pe[pos, i] = np.sin(pos / (10000 ** (2 * i / d_model)))
            if i + 1 < d_model:  # i가 index를 넘어가는 것을 방지
                pe[pos, i + 1] = np.cos(pos / (10000 ** (2 * i / d_model)))
    
    return pe.astype(np.float32)
```

```python
train_frames = []
test_frames = []
```

```python
for building_num in total_data['건물번호'].unique():
    building_data = total_data[total_data['건물번호'] == building_num]
    
    # 20220601 00부터 20220824 23까지 train data로 저장
    train_frames.append(building_data[(building_data['일시'] >= '2022060100') & (building_data['일시'] <= '2022082423')])
    
    # 20220845 00부터 20220831 23까지 test data로 저장
    test_frames.append(building_data[(building_data['일시'] >= '2022082400') & (building_data['일시'] <= '2022083123')])

train_data = pd.concat(train_frames)
test_data = pd.concat(test_frames)
```
